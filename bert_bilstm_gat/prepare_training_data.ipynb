```python
import os
import shutil

import pandas as pd 
import numpy as np 

import pathlib
from sklearn.model_selection import train_test_split
from collections import Counter
```

```python
raw_data_train_path = 'data/fulltrain.csv'
# raw_data_test_path = 'data/balancedtest.csv'

df_train = pd.read_csv(raw_data_train_path, header=None)
# df_test = pd.read_csv(raw_data_test_path, header=None)
df_train.head()
```

```python
df_train.dropna(axis=0, inplace=True)

X_train= np.array(df_train[1]).reshape((-1,1))
y_train = np.array(df_train[0]).reshape((-1,1))
# X_test= np.array(df_test[1])
# y_test= np.array(df_test[0])
```

```python
Counter(y_train.flatten())
```

```python
from imblearn.under_sampling import RandomUnderSampler 

rus = RandomUnderSampler(random_state=123)
X_train, y_train = rus.fit_resample(X_train, y_train)
Counter(y_train.flatten())
```
